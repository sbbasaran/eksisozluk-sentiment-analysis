```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json
from tqdm import tqdm
import pandas as pd
import json

from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
```

```python
model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
```

```python
df = pd.read_csv("/Users/ideakadikoy/Desktop/entries_processed.csv")
```

```python
df=df.dropna(subset=["Processed_Entry"]).reset_index(drop=True)
```

```python
df["Processed_Entry"].apply(lambda x: len(x))
```

```python
df["Entry_Length"] = df["Processed_Entry"].apply(lambda x: len(x))
```

```python
df = df[df["Entry_Length"] < 512].reset_index(drop=True)
```

```python
for i in range(len(df)):
    if i % 100 == 0:
        print(f"{i}/{len(df)} | {i/len(df)*100:.2f}%")
    df.loc[i, "sentiment"] = sa(df.loc[i, "Processed_Entry"])
    
print(f"{len(df)}/{len(df)} | {len(df)/len(df)*100:.2f}%")
```
